# **Basic Plots**

```python
import requests
url = "https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
response = requests.get(url)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from io import StringIO
import matplotlib.pyplot as plt
```

```python
data = StringIO(response.text)
data = pd.read_csv(data)
print("Number of data points =", len(data))
data.head()
```

```python
data.info()
```

```python
data.replace("?", np.nan, inplace = True)
data = data.dropna()
print(len(data))
```

```python
X = data.drop("price", axis = 1)
y = pd.to_numeric(data["price"])
```

```python
plt.figure(figsize = (24, 7))
plt.scatter(X["make"], y)
plt.xlabel('Make')
plt.ylabel('Price')
plt.title('Car maker vs Price - Scatter Plot')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(24,7)})
sns.boxplot(x=X["make"],y=y, palette="Set3").set_title('Car Manufacturer vs Price - Box Plot')
```

**1)Think about the maximum/minimum values for each body-type, which body-type is the "costliest" ?**

**2)Do you think maximum/minimum values are a clear indication of which body-type is costliest, or does the interpretation improve by also taking the interquartile range into consideration?**

The maximum and minimum values alone may not provide a complete or accurate representation of the costliest body type, as they can be influenced by outliers or extreme values. The interquartile range (IQR), on the other hand, is a measure of statistical dispersion that considers the range within which the central 50% of the data falls. Including the IQR in your analysis can provide a more robust understanding of the distribution of values.

When assessing the costliest body type, you should consider both the central tendency (mean or median) and the spread of the data (range, standard deviation, or IQR). The mean is sensitive to outliers, so if there are extreme values in the data, the median might be a more robust measure of central tendency. Additionally, examining the IQR helps identify the range where the majority of the data points lie, making it less influenced by extreme values.

In summary, while maximum and minimum values offer a glimpse into the data range, incorporating measures like the interquartile range provides a more nuanced understanding of the distribution and can improve the interpretation of which body type is costliest.
